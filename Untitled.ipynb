<a href="https://numpy.org/doc/stable/reference/generated/numpy.linspace.html">linspace</a>

```python
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
# col1 is create using linspace wiyh 10 vars
col1 = np.linspace(0, 10, 1000)

# col2 is a collection of col1 sin vars  
col2 = np.sin(col1) 

# DataFrame({ColName1:Vale1, ColName2:Value2,...}) function create dataframe 
df = pd.DataFrame({'C1':col1, 'C2':col2})

df.head(10)
```

```python
# using matplotlib and seaborn 
plt.style.use('seaborn-darkgrid')
%matplotlib inline
sns.lineplot(x=df.C1, y=df.C2, data=df)
plt.show
```

```python
""" - Adjusting background color using axes.facecolor
    - Changing label size using xtick.labelsize, ytick.labelsize"""

plt.figure(figsize=(14,6))
sns.set(rc={"axes.facecolor":"#FFFDE7", "axes.grid":True, 'xtick.labelsize':14, 'ytick.labelsize':6})
sns.lineplot(x=df.C1, y=df.C2, data=df, linewidth=2)
sns.lineplot(x=df.C1, y=df.C2+1, data=df, linewidth=2)
plt.show()
```

```python
""" - Adjusting background color using axes.facecolor
    - Changing label size using xtick.labelsize, ytick.labelsize"""

plt.figure(figsize=(14,6))
sns.set(rc={"axes.facecolor":"#283747", "axes.grid":False, 'xtick.labelsize':14, 'ytick.labelsize':6})
sns.lineplot(x=df.C1, y=df.C2, data=df, color ="#FF5722",linewidth=2)
sns.lineplot(x=df.C1, y=np.cos(df.C1) , color="#FFEB2B", data=df, linewidth=2)
plt.show()
```
